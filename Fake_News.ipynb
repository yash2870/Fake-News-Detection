```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# **Data** **Collection**

```python
true=pd.read_csv('True.csv')
```

```python
fake=pd.read_csv('Fake.csv')
```

```python
true.head()
```

```python
fake.head()
```

```python
true['label']=1
```

```python
fake['label']=0
```

```python
true.head()
```

# **Data Preprocessing**

```python
news=pd.concat([fake,true],axis=0)
```

```python
news.head()
```

```python
news.tail()
```

```python
news.isnull().sum()
```

```python
news = news.drop(["title", "subject","date"], axis = 1)
```

```python
news.head()
```

```python
news=news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'] ,axis=1, inplace=True)
```

```python
news.head()

```

```python
import re
```

```python
def wordopt(text):
  text=text.lower()
  text=re.sub(r'http?:??://\S+[www\.\S+]','',text)
  text=re.sub(r'<.*?>','',text)
  text=re.sub(r'[^\w\S]','',text)
  text=re.sub(r'\d','',text)
  text=re.sub(r'\n','',text)
  return text
```

```python
news['text']=news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x=news['text']
y=news['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization=TfidfVectorizer()
```

```python
xv_train=vectorization.fit_transform(x_train)
```

```python
xv_test=vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

# **Model Creation**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(xv_train,y_train)
```

```python
pred=lr.predict(xv_test)
```

```python
lr.score(xv_test,y_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(xv_train,y_train)
```

```python
pred_dt=dt.predict(xv_test)
```

```python
dt.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
pred_rfc=rfc.predict(xv_test)
```

```python
rfc.fit(xv_train,y_train)
```

```python
rfc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_rfc))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc=GradientBoostingClassifier()
```

```python
pred_gbc=gbc.predict(xv_test)
```

```python
gbc.fit(xv_train,y_train)
```

```python
gbc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_gbc))
```

```python
def output_label(n):
  if n==0:
    print("It is a fake news")
  elif n==1:
    print("It is not a fake news")
```

```python
def manual_testing(news):
  testing_news={"text":[news]}
  new_def_test=pd.DataFrame(testing_news)
  new_def_test['text']=new_def_test["text"].apply(wordopt)
  new_x_test=new_def_test["text"]
  new_xv_test=vectorization.transform(new_x_test)
  pred=lr.predict(new_xv_test)
  pred_gbc=gbc.predict(new_xv_test)
  pred_rfc=rfc.predict(new_xv_test)
  return "\n\nLR Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_label(pred[0]),output_label(pred_gbc[0]),output_label(pred_rfc[0]))
```

```python
news_article=str(input())
```

```python
manual_testing(news_article)
```
